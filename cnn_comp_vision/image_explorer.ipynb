# Explore images

display just some few random images from a folder. Less than 10 images.

We do this, because the folder has more than 20,000 images, and the graphic file explorer get hang for ever...

```python
import os
import random

import matplotlib.pyplot as plt
import matplotlib.image  as mpimg
```

```python
src_dir = '/home/art/data/face_age/utkface'
```

```python
file_names = os.listdir( src_dir )
```

```python
print( 'number of files: {}'.format( len(file_names) ) )
```

```python
print( 'first element: ', file_names[0] )
```

```python
# shuffle the list

random.shuffle( file_names )
```

### function that plot 12 random images

```python
def plot_grid_of_pix( src_dir,
                      file_names,
                      num_rows =4, num_cols = 4, size_inches=4 ):
    
    fig = plt.gcf()
    fig.set_size_inches( num_rows * size_inches, num_cols * size_inches )

    a = [ os.path.join( src_dir, i ) for i in file_names ]

    i = 0
    for img_path in a:
        
        sp = plt.subplot( num_rows, num_cols, i + 1 )
        sp.axis( 'Off' )
        img = mpimg.imread( img_path )
        plt.imshow( img )
        i = i + 1
    plt.show()
```

```python
# get 10 random elements

a = random.sample( file_names, 16 )
```

```python
plot_grid_of_pix(  src_dir, a )
```
