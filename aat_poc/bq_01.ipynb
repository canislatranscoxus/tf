### big query

```
PROJECT = 'my_project'
```

```
import os
os.environ[ 'PROJECT' ] = PROJECT

```

```
%%bigquery --project $PROJECT

select *
from `nyc-tlc.yellow.trips`
limit 10
```

```
from google.cloud import bigquery

bq = bigquery.Client( project = PROJECT )
dataset = bigquery.Dataset( bq.dataset( 'taxifare' ) )

try:
  bq.create_dataset( dataset )
  print( 'taxifare dataset created!' )
except:
  print( 'dataset already exists' )

```

### split data

* 80 %  training
* 10 %  eval
* 10 %  testing



```
%%bigquery 
CREATE OR REPLACE my_dataset.trainig AS

select *
from `nyc-tlc.yellow.trips`
where 
      ABS(
      MOD(
      FARM_FINGERPRINT( CAST( pickup_datetime as STRING )  )
      , 100
      )) < 8




```

```
%%bigquery
CREATE OR REPLACE my_dataset.eval AS

select *
from `nyc-tlc.yellow.trips`
where 
      ABS( MOD( FARM_FINGERPRINT( CAST( pickup_datetime as STRING )  ), 100 )) = 8 
```

```
%%bigquery
CREATE OR REPLACE my_dataset.eval AS

select *
from `nyc-tlc.yellow.trips`
where 
      ABS(MOD( FARM_FINGERPRINT( CAST( pickup_datetime as STRING )  ), 100 )) = 9
```
